```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARIMA
TempData=pd.read_csv('HCLTECH.csv')
TempData.head(30)
```

```python
#Data cleaning
stockdata=TempData.dropna()

stockdata.index=pd.to_datetime(stockdata.Date)
stockdata=stockdata["Prev Close"]['2013-01-01':'2013-12-2']
stockdata.describe()
```

```python
#EDA 
plt.figure(figsize=(16,9))
fig=plt.figure(1)
ax1=fig.add_subplot(111)
ax1.set_xlabel('Time Frame')
ax1.set_ylabel('Stock Price for HCLTECH')
ax1.plot(stockdata)
```

```python
#Checking stationarity
```

```python
rmean=stockdata.rolling(12).mean()
rstd=stockdata.rolling(12).std()

plt.figure(figsize=(16,9))
fig=plt.figure(1)

orig=plt.plot(stockdata,color="blue",label="Original")
mean=plt.plot(rmean,color="red",label="Rolling Mean")
std=plt.plot(rstd,color="black",label="Rolling Std")
plt.legend(loc="best")
plt.title("Rolling Mean and Standard Deviation")
plt.show(block="False")
```

```python
#transformation
plt.figure(figsize=(16,9))
fig=plt.figure(1)

import numpy as np
ts_log=np.log(stockdata)
plt.plot(ts_log)
```

```python
#Decomposition

from scipy.signal.signaltools import residue

from statsmodels.tsa.seasonal import seasonal_decompose
dcmp=seasonal_decompose(ts_log,freq=1,model="mulliplicative")

trend=dcmp.trend
seasonal=dcmp.seasonal
residual=dcmp.resid

plt.figure(figsize=(16,9))
fig=plt.figure(1)

plt.subplot(411)
plt.plot(ts_log,label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
```

```python
#differencing
plt.figure(figsize=(16,9))
fig=plt.figure(1)

ts_log_diff=ts_log-ts_log.shift()
plt.plot(ts_log_diff)

#Rolling statistics
rolLmean=ts_log_diff.rolling(12).mean()
rolLstd=ts_log_diff.rolling(12).std()

orig = plt.plot(ts_log_diff, color='blue',label='Original')
mean = plt.plot(rolLmean, color='red', label='Rolling Mean')
std = plt.plot(rolLstd, color='black', label = 'Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

```python
stockdata.sort_index(inplace=True)
```

```python
from statsmodels.tsa.stattools import acf,pacf
lag_acf = acf(ts_log_diff, nlags=20)
lag_pacf = pacf(ts_log_diff, nlags=20)
```

```python
import statsmodels.api as sm
fig=plt.figure(figsize=(16,9))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ts_log_diff.dropna(),lags=40,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(ts_log_diff.dropna(),lags=40,ax=ax2)
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
type(ts_log_diff)
```

```python
ts_log_diff=ts_log_diff[-ts_log_diff.isnull()]
```

```python
plt.figure(figsize=(16,9))
#ts_log_diff.dropna(inplace=True)
model = ARIMA(ts_log_diff, order=(1,1,1))  
results_ARIMA = model.fit()  
plt.plot(ts_log_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
```

```python
#taking results back to original scale
```

```python
ARIMA_diff_predictions = pd.Series(results_ARIMA.fittedvalues, copy=True)
print(ARIMA_diff_predictions.head())
```

```python
ARIMA_diff_predictions_cumsum = ARIMA_diff_predictions.cumsum()
print(ARIMA_diff_predictions_cumsum.head())
```

```python
ARIMA_log_prediction = pd.Series(ts_log.iloc[0], index=ts_log.index)
ARIMA_log_prediction = ARIMA_log_prediction.add(ARIMA_diff_predictions_cumsum,fill_value=0)
ARIMA_log_prediction.head()
```

```python
plt.figure(figsize=(12,8))
predictions_ARIMA = np.exp(ARIMA_log_prediction)
plt.plot(stockdata)
plt.plot(predictions_ARIMA)
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_ARIMA-stockdata)**2)/len(stockdata)))
```

```python
import pmdarima as pm
def arimamodel(timeseries):
    automodel = pm.auto_arima(timeseries, 
                              start_p=3, 
                              start_q=3,
                              max_p=5,
                              max_q=5,
                              test="adf",
                              seasonal=True,
                              trace=True)
    return automodel
```
